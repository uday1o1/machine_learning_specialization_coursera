```python
mcu_box_office_df.columns = [col.replace(' ', '_') for col in mcu_box_office_df.columns]
```

```python
mcu_box_office_df.Release_Date = pd.to_datetime(mcu_box_office_df.Release_Date)
```

```python
mcu_box_office_df.Release_Date = mcu_box_office_df.Release_Date.dt.strftime('%d/%m/%Y')
```

```python
mcu_box_office_df.info()
```

```python
mcu_box_office_df = mcu_box_office_df.rename(columns={"Movie": "Title", "Inflation_Adjusted_Domestic": "Revenue_US", 
                                  "Inflation_Adjusted_International": "Revenue_International", 
                                  "Inflation_Adjusted_Worldwide": "Revenue_Worldwide", 
                                  "Inflation_Adjusted_Opening_Weekend": "Revenue_Opening_Weekend", 
                                  "Inflation_Adjusted_Budget": "Budget"})
```

```python
mcu_box_office_df["Month"] = pd.DatetimeIndex(mcu_box_office_df.Release_Date).month
mcu_box_office_df["Year"] = pd.DatetimeIndex(mcu_box_office_df.Release_Date).year
```

```python
mcu_box_office_df.info()
```

```python
data = mcu_box_office_df.groupby("Year").count()["Title"]
print(data.tail())

mcu_box_office_df.groupby('Year').count()['Title'].plot(xticks = np.arange(1985,2022,5))

sns.set(rc={'figure.figsize':(10,5)})
plt.title('Year vs number of Movies', fontsize = 15)
plt.xlabel('Release year', fontsize = 13)
plt.ylabel('Number of Movies', fontsize = 13)

sns.set_style('whitegrid')
```

```python
def find_minmax(x):
    min_index = mcu_box_office_df[x].idxmin()
    high_index = mcu_box_office_df[x].idxmax()
    high = pd.DataFrame(mcu_box_office_df.loc[high_index,:])
    low = pd.DataFrame(mcu_box_office_df.loc[min_index,:])
    
    print('Movie Which had the Highest '+ x +':',mcu_box_office_df['Title'][high_index])
    print('Movie Which had the Lowest ' + x +':',mcu_box_office_df['Title'][min_index])
    print()
    return pd.concat([high,low],axis = 1)
```

```python
find_minmax('Revenue_Worldwide')
```

```python
info = pd.DataFrame(mcu_box_office_df['Revenue_Worldwide'].sort_values(ascending = False))
info['Title'] = mcu_box_office_df['Title']
data = list(map(str,(info['Title'])))
x = list(data[:10])
y = list(info['Revenue_Worldwide'][:10])

ax = sns.pointplot(x=y,y=x)

sns.set(rc={'figure.figsize': (10,5)})
ax.set_title('Top 10 Highest Grossing Marvel Movies Worldwide', fontsize = 25)
ax.set_xlabel('Earnings',fontsize = 15)
sns.set_style('darkgrid')
```

```python
info = pd.DataFrame(mcu_box_office_df['Budget'].sort_values(ascending = False))
info['Title'] = mcu_box_office_df['Title']
data = list(map(str,(info['Title'])))

x = list(data[:10])
y = list(info['Budget'][:10])

ax = sns.pointplot(x=y,y=x)
sns.set(rc={'figure.figsize':(10,5)})
ax.set_title('Top 10 Highest Budget Films', fontsize = 25)
ax.set_xlabel('Budget', fontsize = 15)
sns.set_style('darkgrid')
```

```python
def count_prod(x):
    data_plot = mcu_box_office_df[x].str.cat(sep = '|')
    data = pd.Series(data_plot.split('|'))
    info = data.value_counts(ascending= False)
    return info
```

```python
production_companies = count_prod('Ownership')
production_companies.iloc[:10].plot(kind='barh',figsize=(16,8),fontsize=15)
plt.title('Ownership vs Number of Movies',fontsize = 30)
plt.xlabel('Number of Movies',fontsize = 15)
sns.set_style('whitegrid')
```
